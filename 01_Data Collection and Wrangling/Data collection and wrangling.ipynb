
We are going to format the ‘REGION’ column: we will leave only the area values 1,2,3,4,5 and 6 and we will remove the string ‘TOTAL AUTOS AREA’. In addition, the records named ‘TOTAL AUTOS SCANNING MEXICO’ will be assigned a 0. This will be done with the intention of optimizing the DF when machine learning algorithms are applied later.


```python
# Definicion de funcion para realizar el cambio en la columna REGION
def cambio_numero (palabra):
    if palabra == 'TOTAL AUTOS SCANNING MEXICO':
        return 0
    elif palabra == 'TOTAL AUTOS AREA 1':
        return 1
    elif palabra == 'TOTAL AUTOS AREA 2':
        return 2
    elif palabra == 'TOTAL AUTOS AREA 3':
        return 3
    elif palabra == 'TOTAL AUTOS AREA 4':
        return 4
    elif palabra == 'TOTAL AUTOS AREA 5':
        return 5
    elif palabra == 'TOTAL AUTOS AREA 6':
        return 6

# Aplicacion de la funcion de en la columna REGION
fact_sales_df['REGION'] = fact_sales_df['REGION'].apply(cambio_numero)
fact_sales_df['REGION'].value_counts()
```

# Part 3: data frame merging

In this section we join dataframes to consolidate the information and have less number of individual tables. First, we will merge the dataframes fact_sales_df and dim_product_df, which have in common one code per item (‘ITEM_CODE’ and ‘ITEM’ columns, respectively). This will be done with the merge() function.

```python
# dataframe merging of fact_sales_df and dim_product_df using the merge() method
productos_vendidos = pd.merge(fact_sales_df, dim_product_df, left_on='ITEM_CODE', right_on='ITEM')

# Let's drop the ITEM column  since we already have ITEM_CODE
productos_vendidos.drop('ITEM', inplace=True, axis=1)

# Now we will rename the CATEGORY ID column
productos_vendidos.rename(columns = {'CATEGORY': 'CATEGORY_ID'}, inplace = True)

# Visualize dataframe productos_vendidos with head()
productos_vendidos.head()

# Cambio de nombre del df productos_vendidos
productos_vendidos.rename(columns = {'CATEGORY': 'CATEGORY_ID'}, inplace = True)
productos_vendidos.head()
```

Since we have merged dataframes fact_sales_df and dim_product_df into one single dataframe, we are now going to merge the dim_category_df using the CATEGORY_ID column

```python
# Let's merge productos_vendidos and dim_category_df using CATEGORY_ID column
productos_vendidos = pd.merge(productos_vendidos, dim_category_df, how = 'left', left_on='CATEGORY_ID', right_on= 'ID_CATEGORY')

# Eliminacion of the  CATEGORY_ID oclumn
productos_vendidos.drop('CATEGORY_ID', inplace = True, axis = 1)

# Renaming the column 'SEMANA' for 'WEEK'
productos_vendidos.rename(columns={'WEEK' : 'SEMANA'}, inplace = True)

# Finally, we will add the date to the s merging 
productos_vendidos = pd.merge(productos_vendidos, dim_calendar_df, how = 'left', left_on='SEMANA', right_on= 'WEEK')

# Eliminacion de la columna repetida 'WEEK'
productos_vendidos.drop('SEMANA', inplace = True, axis = 1)
```
